# Data cleaning (Entsoe_data)

```python
def clean_data_1(df):
  
    data = df.copy()
  
    # Drop duplicated data
    data = data.drop_duplicates()
    
    # If there are more than 10000 NaN values in a column, drop the column
    for i in data.columns:
        if data[i].isna().sum() > 10000:
            data = data.drop([i], axis=1)
   
    # Fill nan values with an interpolated value
    for i in data.columns:
        data[i] = data[i].fillna((data[i].interpolate()))
    
    # Drop the remaining NaN values (that couldn't be interpolated)
    data = data.dropna()    
        
    # Find the outliers using Inter Quartile Range (IQR) method
    Q1 = data.quantile(q=0.25)
    Q3 = data.quantile(q=0.75)
    IQR = Q3 - Q1
    
    upper_limit = Q3 + 1.5 * IQR
    lower_limit = Q1 - 1.5 * IQR

    # Find and replace the outliers with an interpolated value  
    
    """ I couldn't make it to directly replace the outlier with the interpolated value, 
        so first I replaced them with NaN values then replaced the NaN values with an 
        interpolated value.
    """
    for i in data.columns:
        if lower_limit[i] < 0:
            data.loc[data[i] > upper_limit[i], i] = np.nan
        else:
            data.loc[data[i] > upper_limit[i], i] = np.nan
            data.loc[data[i] < lower_limit[i], i] = np.nan
        
        data[i] = data[i].fillna((data[i].interpolate()))
        
    # Drop the remaining NaN values (that couldn't be interpolated)
        data = data.dropna()
       
    return data
```

# Creating the SQL database

```python

def create_connection(db_file):
    """ create a database connection to a SQLite database """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        print(sqlite3.version)
    except Error as e:
        print(e)
    finally:
        if conn:
            conn.close()

if __name__ == '__main__':
    create_connection(r"clean_database.db")
```
